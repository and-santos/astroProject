```python
# Importações iniciais

from astropy.io import fits
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Detalhes sobre as colunas do arquivo fits
https://www.legacysurvey.org/dr9/catalogs/#region-tractor-aaa-tractor-brick-fits

```python
hdul = fits.open("./catalogs/sweep-000m005-010p000.fits", ignore_missing_end=True) # importando um arquivo fits
data = hdul[1].data  # table extensions can't be the first extension, so there's a dummy image extension at 1
data_table = Table(data)
data_table

# Pandas don't support multidimensinal columns
data_table.remove_column('DCHISQ')
df = data_table.to_pandas()  # Erro na conversão devido a coluna multidimensional DCHISQ
```
