## Manipulação de pandas.DataFrame

### Importando bibliotecas

```python
from astropy.io import fits
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

### Open catalog from fits table using astropy package

```python
hdul = fits.open('legacy_dr8_class.fits')
hdul.info()  # Information about openend fits catalog, BinTableHDU is in hdul[1].
```

### Fits Table and columns

```python
data = hdul[1].data  # table extensions can't be the first extension, so there's a dummy image extension at 1
cat_table = Table(data)
cols = cat_table.columns
print(cols)
cat_table
```

### Constructing pandas.DataFrame from Table fits

```python
df = cat_table.to_pandas()
df
```

### Add column to pandas.DataFrame

pandas.DataFrame.astype(dtype, copy=True, errors='raise'): Cast a pandas object to a specified dtype dtype.

```python
new_col = '_id'  # name new column
df['ra'] = df['ra'].astype(str)  # change numbers to string
df['dec'] = df['dec'].astype(str)  # change numbers to string
df[new_col] = df['ra'] + df['dec']  # join two columns
df
```

### Splitting in two pandas.DataFrame

```python
print("DataFrame 1")
df_1 = df.iloc[:, 2:-7]  # set mags columns dataframe 1 from dataframe
df_1[new_col] = df[new_col]  # inset Id column

# The lines below pushes '_id' to the beginning of list
cols1 = df_1.columns.tolist()  # convert pandas.Series to list
cols1 = cols1[-1:] + cols1[:-1]  #cols1[-1:] -> catches only the '_id' column, cols1[:-1] -> catches the other ones
df_1 = df_1[cols1]  # set first column _id in dataframes 1

print(df_1)
print(df_1.columns)

print('******************************')
print("DataFrame 2")

df_2 = df.iloc[:,-7:-2]  # set photoz columns dataframe 2 from dataframe
df_2[new_col] = df[new_col]
cols2 = df_2.columns.tolist()
cols2 = cols2[-1:] + cols2[:-1]  # set first column _id
df_2 = df_2[cols2]  # set first column _id
print(df_2)
print(cols2)
```

```python
df_1 = pd.read_csv(os.path.join(location, 'LegacyCatalog1.csv'))
df_2 = pd.read_csv(os.path.join(location, 'LegacyCatalog2.csv'))
```

### Using the _id column as a reference finally we can now perfomance the join operation

```python
DF_mag = df_1.set_index(new_col)
DF = df_2.join(DF_mag, on=new_col)
DF
```

### Saving and loading new dataframe in table fits

```python
t2 = Table.from_pandas(DF)
fits_table_filename = os.path.join(location, 'NewLegacyCatalog.fits')
hdu = fits.BinTableHDU(data=t2)
hdu.writeto(fits_table_filename, overwrite=True)

hdul = fits.open(fits_table_filename, ignore_missing_end=True) 
data = hdul[1].data # table extensions can't be the first extension, so there's a dummy image extension at 1
cat_table = Table(data)
cols=cat_table.columns
print(cat_table)
print(cols)
```
